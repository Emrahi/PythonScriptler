```python
import pandas as pd
from selenium import webdriver

page_url = "http://rpachallenge.com/"
excel_url = page_url + "assets/downloadFiles/challenge.xlsx"
```

```python
data=pd.read_excel(excel_url,usecols=range(7), names=["fname","lname","comp","role","addr","email","phone"])
```

```python
data
```

```python
driver = webdriver.Chrome()
driver.get(page_url)
```

```python
#ba≈ülat
driver.find_element_by_xpath("//button").click()
```

```python
for i in range(len(data)):
        driver.find_element_by_xpath("//LABEL[@_ngcontent-c2=''][text()='First Name']/following-sibling::INPUT").send_keys(data.fname[i])
        driver.find_element_by_xpath("//LABEL[@_ngcontent-c2=''][text()='Last Name']/following-sibling::INPUT").send_keys(data.lname[i])
        driver.find_element_by_xpath("//LABEL[@_ngcontent-c2=''][text()='Company Name']/following-sibling::INPUT").send_keys(data.comp[i])
        driver.find_element_by_xpath("//LABEL[@_ngcontent-c2=''][text()='Role in Company']/following-sibling::INPUT").send_keys(data.role[i])
        driver.find_element_by_xpath("//LABEL[@_ngcontent-c2=''][text()='Address']/following-sibling::INPUT").send_keys(data.addr[i])
        driver.find_element_by_xpath("//LABEL[@_ngcontent-c2=''][text()='Email']/following-sibling::INPUT").send_keys(data.email[i])
        driver.find_element_by_xpath("//LABEL[@_ngcontent-c2=''][text()='Phone Number']/following-sibling::INPUT").send_keys(str(data.phone[i]))

        driver.find_element_by_xpath("//input[@value='Submit']").click()
        
    
    
```
